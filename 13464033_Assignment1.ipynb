## Content

Face Recognition is one of the most challenging tasks in the image analysis and pattern recognition field wherein the samples of the images are taken, classified, and compared with other images in the database (Sukhija, Behal & Singh 2016). The face is a deformable object, and expressions in faces are available in a substantial form of viable configurations. Recognition of face is done by comparing the features which represent the variation in face structure, shape and different face attributes like distance between the eye and node and eyebrow width and the accuracy of the classification is expected to be high. Faces tend to change over years, and several features such as wrinkles, skin color and pores sagging are to be eliminated. The entire face recognition process is influenced by several factors such as pose, light illumination, reflectance, and shape, thus making it more difficult to predict (Bhatia, Lilhore & Agrawal 2017).

Several algorithms have been introduced to solve the problem of face detection and enhancing the accuracy of the expirement.The first face recognition algorithm was introduced in the 1960's wherein the human face was detected using the geometric features.In 1991, Turk expanded the concept of Eigenfaces on identifying how to recognize the faces within images which led to the introduction of automatic face recognition.

Few Challenges while during the face recognition are as follows:
* The images that are taken vary such as tilted, upside down every time due to the camera-face pose.
* When an image is taken, factors such as lightning, pose and camera’s position tends to differ each time that in turn makes it difficult to make the prediction
* When taking pictures, a person from the same viewpoint and expression appears different under various lighting conditions.Images that are not taken years ago may not match with the pictures in a database that makes it difficult to recognize.

This paper attempts to provide a more error-free face detection and recognition method from various images which are not sensitive to huge variations in facial expressions and lighting from multiple sources.

The idea of the paper is to linearly project an image into a subspace which does not deviate more from its path and not produce self-shadowing. Various linear projection algorithms such as Fisherface, Eigenface were used 

in such a way that projected images in a low dimensional space under severe variation in lightning and facial expressions. Projections are made orthogonal to the class scatter by differentiating the variations in lighting and facial expressions.
