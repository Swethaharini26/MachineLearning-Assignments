## Introduction: 


The selected paper, as it goes by the topic, is about a machine learning algorithm called Linear Discriminant Analysis (LDA) /Principle Component Analysis (PCA) and illustrates the effective way of selecting the features, transforming the data to a high dimensional space and using the algorithms on a selected dataset to perform face recognition. Linear Discriminant Analysis and Principle Component Analysis are linear transformation techniques where in LDA is a supervised learning algorithm whereas PCA ignores the class label. However, the PCA is generally used in dimensionality reduction but not as a classification method since there will be no correlation between the features when applied on data, instead it can be used to improve the performance of LDA. The idea described in this paper is to develop a face recognition algorithm that is not sensitive to the variation in the direction of the lightning, number of light resources and facial expressions. The author has attempted to solve the face recognition problem using several pattern classification techniques namely Correlation, Eigenfaces, Linear subspaces and Fisherspaces and provided a comparative study on the algorithms by modifiying the features to identify the effective feature set and algorithm to solve the face recognition problem.


## Content

Face Recognition is one of the most challenging tasks in the image analysis and pattern recognition field wherein the samples of the images are taken, classified and compared with other images in the database. (Pratibha Sukhija, et al., 2016). The face is a deformable object, and expressions in faces are available in a substantial form of viable configurations. Recognition of face is done by comparing the features which represents the variation in face structure, shape and different face attributes like distance between the eye and node and eyebrow width and the accuracy of the classification is expected to be high. Faces tend to change over years and several features such as wrinkles, skin color and pores sagging are to be eliminated. The entire face recognition process is influenced by several factors such as pose, light illumination, reflectance and shape, thus making it more difficult to predict. 

Few Challenges while during the face recognition are as follows:

-	The images that are taken vary such as tilted, upside down every time due to the camera-face pose.
-	When an image is taken, factors such as lightning, pose and camera’s position tends to  differs each time that in turn makes it difficult to do the prediction
-	When taking pictures, a person from the same view point and expression appears different under various lightning conditions.
-	Images that are not taken years ago may not match with the images in database that makes it difficult to recognize.

Several algorithms have been introduced to solve the problem of detecting the faces and enhancing the accuracy of the recognized faces. The first face recognition algorithm was introduced in 1960’s wherein the human face was detected using the geometric features. In 1991, Turk expanded the concept of Eigen faces on identifying how to detect the faces within images which led to the introduction of automatic face recognition.

This paper attempts to provide a more error free face detection and recognition method from various images which are not sensitive to huge variations in facial expressions and lightning from various sources. The idea of the paper is to linearly project an image into a subspace which does not deviate more from its path and not produce self-shadowing. To achieve this, various linear projection algorithms such as Fisher face, Eigen face were used that projected images in a low dimensional space under severe variation in lightning and facial expressions. Projections are made orthogonal to the class scatter by differentiating the variations in lightning and facial expressions.

The paper is divided into various sections wherein the first section of the paper deals with the problems while during the face detection and how the detection can be made easier. To solve this problem, four methods were compared under various lightning conditions and determined which method can be used to make an accurate and error- free prediction. In the second section, assuming that the image is in high dimensional space, a training set with the labeled set of images and a test set that is not labeled is taken to do the comparison.
Considering the hypotheses to be tested and the expected performance from each algorithm, the data that has variations in lightning are selected. In the third section, experiments are conducted under variations in lightning, facial expression and eye wear. Comparative Study is done to determine if the person is wearing glasses and identifying the error rate with Eigen faces and Fisher faces algorithms.

Overall, this paper provides a comparative study on using Eigen faces (Principle Component Analysis) and Fisher faces (Linear Discriminant Analysis) for face recognition under various factor such as lightning conditions, the angle the image is to be projected, facial features and the projected surface.


## Innovation

Over the years many algorithms have proposed to solve the problem of human face recognition and achieve accurate results. Classification of data and dimensionality reduction can be done using LDA and PCA. Tests prove that LDA shows outstanding performance when tested on a given set of data, provided the within-class frequencies are unequal. PCA focuses more on feature recognition whereas LDA focuses more on data classification, thereby, trying to provide more class separability and arrives at a conclusion between the given classes.

The goal of the paper is to develop a face recognition algorithm that is not sensitive to light variations and facial expressions. The formation of the paper is very good as the author has provided a comparative study on various algorithms that can be used for face recognition by addressing the drawbacks for each rather concluding the best method for performing face detection. The author begins the study by stating that, since the images produce self-shadowing and are not truly Lambertian surface, the image should be linearly projected into a subspace which discounts all the regions that produce large deviation. To make the linear projection to a low dimensional subspace, Fisher’s linear Discriminant and Eigenfaces techniques are used. The author has taken two databases to perform the prediction; Harvard and Yale Face databases. He has not considered the pose variations of a person to be one of his characteristics for prediction as it may be difficult to analyze under different lightning conditions. To choose the best method, he has provided a comparative study on four different algorithms for face detection that are as follows:

* Correlation
* Eigenface method
* Fisherface method 
* Linear subspaces

The author has attempted to detect the face of a person using correlation method wherein the labels from the test set are assigned to images in the training test based on nearest neighbor algorithm. He has identified that the normalization process is independent of light sources wherein there would not exist clustering if the image in test and training set are considered under various lightning conditions. One major disadvantage addressed in the paper is the time taken to correlate an image in the training test to test set which would reduce the computation time. The author has also addressed the cost and storage as one of the most important factors for not considering this method to do the recognition. 

The next method which is discussed in the paper is Eigenfaces for performing face recognition wherein the image is linearly projected into a linear sub space. The basic idea is to calculate the eigenvectors of the covariance matrix, and total the original data by a linear combination of the leading eigenvectors. The author has made a detailed study on how the eigen vectors can be obtained with the given samples. The Eigenface method uses the Principle Component Analysis (PCA) for dimensionality reduction which results in projection across all the classes. In the PCA-based face recognition technique, 1D image vectors are obtained from the transformation of 2D images.This, inturn leads to the complexity of computing the covariance matrix due to the large size. However, the eigenvectors are determined by the covariance matrix and accuracy is not sustained when approaches change.

The major drawback here is since the scatter across all classes is high, PCA retains the unwanted variations that occur due to lightning and facial expressions which result in poor clustering among the projected point across a linear subspace.
