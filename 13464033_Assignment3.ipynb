ATTENDANCE MONITORING SYSTEM FOR UTS

AIM :
  
  This project is dedicated to is to develop a system for the University of Technology Sydney, which tracks the attendance of the students that attends a particular lecturer. This system generally tracks the students who attend the lecture and sends an email to the professor a list of students who have attended or skipped a particular lecture. Additionally, this system also automatically sends an email to the students for declaring the reason for their absence and stores it to the database of the lecturers. The system is purposely designed to be limited to three warnings for the students. If a student is found to miss the lecture more than three times without any proper response, system will be able to immediately retrieve the student's detail and escalate to Head of UTS after the lecture decided it. To make this system user-friendly, UI will be developed for the lecturer to upload the images of the classroom and the system marks the attendance for that session by itself. This system minimizes the efforts of taking the attendance manually in excel sheet and would be helpful to monitor the presence of students in a classroom in a nonintrusive manner using Machine learning technique called Face Recognition with Convolutional Neural Networks (Kasturi & Manchukonda 2018).

What is Face Recognition?

Face Recognition is the ability to uniquely identify the identity of a person by analyzing and comparing the various images and generating a pattern for accurate identification. The nodal points are set up on a human face, and the values against these points represent the image accurately. By face recognition, it is also clear that it can be used to make liveliness detection to identify the presence of a person.

This technique is divided into two parts: Face Detection and Face Recognition. Face Detection is a technique in which the training data would be given to the model, and the model identifies the person's face that matches the training data. The later part, Face Recognition is harder than the verification as when it recognizes the image with a very weak verification, chances exist that the system may give false predictions. Hence when using this technique care must be taken that the model is trained well and can provide accurate results.


What are Convolutional Neural Networks?

Convolutional Neural Networks, a network that is made up of neurons are primarily adapted to classify the images based on assigning weights for the images and biases. Every neuron in the network receives inputs, performs a dot product operation, thereby passing through an activation function, and responds with an output.

The aim of this project is broken into a few main objectives that are as follows:

(i) The idea is to implement this technique in a very small classroom, i.e., a class that has 20-30 participants to evaluate the results of the prediction made. To make the prediction, the images of the students needs to be collected.

(ii) The next step is to preprocess the images using grey scaling or image resizing techniques for achieving optimal results in face detection.

(iii) Evaluate the performance of different machine learning classifiers and identify the best classifiers that produce accurate results.

(iv) Validate the developed models and deploy them in the real-time environment to evaluate performance.


BACKGROUND:

INTRODUCTION:

Nowadays, education has become the most significant factor in surviving in this world, and people thrive on getting a quality education. The education sectors and the universities are focusing on improving the quality of the content that will be delivered in each lecture, rather concentrating on the number of students attending the classes. On the other hand, the benefits are high only if the students attend the lecture and show some productive results. At present, attendance is being taken by calling out the student's name or passing a sheet wherein the students write their names to declare their presence for a particular lecture.The methods, as mentioned above, have their drawbacks. The calling out of student's name is time-consuming and not practically possible when the size of the classroom is big, and a lot of time may get wasted for taking attendance, and very little time would be left for the lecture. In the second method, where the students writing their names in a paper to mark their presence is usually prone to fraudulent activities wherein a student may coverup his/her friend's absence for a particular lecture. Additionally, this may also lead to inaccuracy (Agrawal et al. 2020).

Lecturers have to monitor all the student's attendance to the class and updates in the system at the end of every lecture. Our system eliminates the traditional recording activity of the professor marking attendance on the paper or excel sheet, calling out the student's names in class.

This system captures the images of the students to identify whether a student has attended a particular lecture or not, thereby marking the attendance automatically and emails the attendance to the professor every day once the lecture is over. With this system, the efforts and pain of tracking student's individually can be eliminated. This system also emails the students asking the output of a particular class and inquires the reason for their absence through email in case if a student has not attended a particular lecture (Sotonwa & Oyeniran 2019).


Current Problems in the Attendance System:

- The time-consuming process as the majority of time is spent on monitoring the student's entry and exit for a particular class.

- There exists a chance that lecturers may mark the attendance wrongly or miss marking the attendance for a student.

- There are chances of getting inconsistencies in data when the attendance has to be marked up for a large number of students.

- Separate training has to be provided to staff on how to update the attendance and upload it in the database.

- Security is the biggest concern while updating the attendance manually as there are high chances of data loss or leakage to occur.

- The manual attendance tracking process is tedious as lecturers find it difficult to identify the students that missed classes and collect the reason for their absence.

The attendance monitoring is currently done manually, and the lecturers are facing difficulties in recording the attendance of every student. Since there exists no proper solution for this, many Universities have removed compulsory attendance for the students. Students take advantage of this liberty and do not attend the classes that, in turn, affect their performances over the years.

Many systems were introduced to automate the process of taking attendance in classrooms and monitoring them. All of the systems had their shortcomings and were not able to produce optimal results. One such technique was to record and monitor the attendance of the students using Fingerprint technology. In this system, the fingerprints of the students are collected and updated in the database. The attendance is marked by matching the existing fingerprint with the fingerprints that were detected. Though this system prevents fraudulent activities, cost of installing the system in every classroom is high, as any campus may have 400-600 classrooms. Additionally, when the size of the classroom is large, and the number of students is high, this technology is a time consuming one. This technique has also not yet provided accurate results, as per the study and so this technique is not considered to be the best fir to record the attendance.

Another technique was proposed using the RFID cards, wherein this kind of card would be issued to the students. The students can tap on their card while entering the classroom and tap off to record the attendance. To implement this system, the entire infrastructure of the building has to be modified that results in high costs. Though the attendance of the students is recorded, lecturers have to meet the student and enter the reason for their absence, which in turn is considered to be one of the major drawbacks of this system.
The problem that is intended to be solved is creating ease in monitoring the student's attendance and updating the reason for their absence to the lecturers in a single system, which is the final output of the proposed project. This is considered to be one of the main reasons for developing this attendance monitoring system.

How does System work?

Initially, the system collects the images of all the students and stores them in a database. Then the images are labeled with the name or id of the corresponding students. With the help of the input images, the model is trained that predicts the name of a person by seeing the image. During the lectures, the images of the student who attend the lectures will be recorded and compare against the samples that are trained. Then, the system marks the attendance for the students who have attended the lecture and automatically sends an email to the students questioning for their absence. The system is designed in such a way that has also set the limit of tracking the unattended students thrice, and if the limit is exceeded, the system will prompt the lecturer to escalate it to the head of university. Here, Text Classification algorithm is used to decide whether the professor will be interested in escalating or not. The model is trained by collecting the previous responses of the lecturer for a similar situation to predict the output. This is the high-level functionality of the attendance monitoring system.


Innovation behind the project:


The innovation behind this project is face recognition using convolution neural networks for training the model. For predicting the lecture's response to the student's absence, the text classification technique is used.Image Recognition is the overall concept of this project. Using this technique, the image can be monitored in real-time, and when the data is trained, algorithms can be used to predict the exact person's name for the images. Additionally classification technique is also used to categorize and classify the text from an email system to predict the response that would be given by the lecturer. Most of the organization are looking for an automation system to record the attendance (Agrawal et al. 2020). The innovation process has been implemented in the system to record the images of a student who attends the classes, and the face recognition technology is used to recognize the faces. Later, once the student responds with the reason for the absence, text classification technique is used to predict the response of the lecturer. The reason for using this technique automates the process as there may occur an emergency when the lecturer would not be able to address the issue that, in turn, may worsen the problem. 

To address a situation like this, the system is incorporated with this mechanism. For example, after the system has been trained with the images, it would be able to identify the names of the person. So, when a professor takes an image of the class, or the camera recordings have uploaded the system, it identifies who has missed and attended the particular lecture. A User interface has been used to serve the purpose of uploading the image, and the system responds by marking the attendance. Then, the system sends an email to the students marked "absent" asking the reason.

Once the responses are received, the system displays it in the lecturer's UI corresponding to the student's name. The lecturer then decides to escalation either manually, or the system can make a self-decision on escalation to the head of the university.

Steps to design a solution for the given problem:


1. Data Acquistion:

Firstly, to train any model or develop any algorithm, data is identified to be a key factor, and the acquired data has to be trained well to get accurate results. For our system, data are the images and the student details which can be collected when a student first enrolls in a subject. A UI is designed for this system with the facility to register before using the system by filling out the basic details and a photograph for identity cards. Those images and the student details are assumed to be samples for training, and those samples are trained to be used with the testing data, which is the photograph of the students updated by the lecturer after a particular class. Care should be taken that images that are to be uploaded in the registration system are to be of high resolution and the lightning for the image, position of the image placed must be appropriate.

2. Acquistion of Lecture and Student Details:

When the student registers into the system, they will be prompted to enter the personal details along with the subjects that they are enrolled for that semester. By doing this, a mapping can be created between the details and the pictures that would be uploaded by the lecturer during class. The system will then be able to classify and identify the person's name that matches the image.

On the other hand, the lecturer should also register by entering the basic details along with the student's name and the classes that are to be handled for that semester.

3. Modelling:

The model can be created using Python, R, or various data visualization tools like Knime, IBM SPSS. Before developing a model, it is very important to do a few preprocessing steps, like enhancing the quality of the image without distorting the important features of the face. The image can be processed according to the amount of light reflected on the face, such as Shadow, dull, or illuminated. Various techniques such as homomorphic filtering can be used to process the image and enhance its quality. Once the pre-processing is done the face detection would be done initially using an algorithm followed by the face recognition wherein the Deep convolutional networks would be used to recognize the different features of the images.

A linear SVM classifier can be trained with the help of the generated face features. The classifier would take the sample image as input, which in turn finds the match for the sample face by trying to match with the face in the database.


4. Deployment:

The performance of the model can be evaluated for very few samples first, and if the model can provide accurate results, then a large number of images can be fed into the model. For that, initially, this system can be implemented to a very small classrooms or conference rooms wherein the count is up to 30 or 40 members. The deployment can be done for two to three weeks continuously by gradually increasing the member count to evaluate performance. 

5. Expected Outcome:

The system should be helpful for the lecturers by reducing their workload and the time spent on marking attendance. A simple application has been created, and the users need not require additional training to use this system.
